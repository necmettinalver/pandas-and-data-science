```python
employee_data = {'employee_id': ['1', '2','3','4'],
'name': ['Mert Cobanov', 'Fethi Tekyaygil', 'Segaf Blackwater','Tim Horton']}

sales_data = {'employee_id': ['3', '4', '5', '6'],
'total_sales': [23456, 2512, 2345, 1455]}
```

```python
df_employee = pd.DataFrame(employee_data)
df_sales = pd.DataFrame(sales_data)
```

```python
concat_employee_df=pd.merge(df_employee,df_sales,on='employee_id',how='outer')
#how, to be left, right. if you are write left or right in how, table The table is shaped accordingly.
```
